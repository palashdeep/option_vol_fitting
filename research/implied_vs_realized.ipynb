```python
import os
import glob
import pandas as pd
from src.viz.plot_utils import plot_iv_vol, plot_smile, plot_iv_error
```

```python
DATA_DIR = os.path.join("data")
DATA_PATH = os.path.join(DATA_DIR, "processed")
```

```python
pattern = os.path.join(DATA_PATH, "vol_data_*.csv")
files = glob.glob(pattern)
if not files:
    print("No files found matching pattern:", pattern)
    print("Please run the data acquisition script to generate the required data files.")
for file in files:
    print("Processing file:", file)
    df = pd.read_csv(file, parse_dates=['date'])
    ticker = df['ticker'].iloc[0]
    print("Ticker:", ticker)
    # Plot IV and Realized Volatility over time
    plot_iv_vol(df, ticker)
    plot_smile(df, '1M', ticker)
    plot_iv_error(df, ticker)
```
